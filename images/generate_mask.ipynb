```python
from plantcv import plantcv as pcv

n = 60
class options:
    def __init__(self):
        self.image = f"./sized/{n}.jpg"
        self.debug = "plot"
        self.writeimg= False
        self.outdir = "."
        
# Get options
args = options()

# Set debug to the global parameter 
pcv.params.debug = args.debug

# Read image

# Inputs:
#   filename - Image file to be read in 
#   mode - How to read in the image; either 'native' (default), 'rgb', 'gray', or 'csv'
img, path, filename = pcv.readimage(filename=args.image)
a = pcv.rgb2gray_lab(rgb_img=img, channel='a')
img_binary = pcv.threshold.binary(gray_img=a, threshold=115, max_value=255, object_type='dark')
id_objects, obj_hierarchy = pcv.find_objects(img=img, mask=img_binary)
obj, mask = pcv.object_composition(img=img, contours=id_objects, hierarchy=obj_hierarchy)
masked = pcv.apply_mask(img=img, mask=mask, mask_color="black")
pcv.print_image(img_binary, f"./mask/{n}.jpg")
```
